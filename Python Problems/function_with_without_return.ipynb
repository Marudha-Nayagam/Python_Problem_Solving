```python
age = int(input("enter age"))
```

```python
# with return
def ageCategory():
    if age < 18:
        print("children")
        cate="children"
    elif age < 39:
        print("adult")
        cate = "adult"
    elif age <59 :
        print("citizen")
        cate = "citizen"
    else:
        print("senior citizen")
        cate = "senior citizen"
    return cate
```

```python
result = ageCategory()
```
