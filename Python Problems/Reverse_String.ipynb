```python
# Using slice  
# ⭐ Fastest method

text = "hello"
reversed_text = text[::-1]
print(reversed_text)
```

```python
# using reversed() function and join
# fast

text = "hello"
reversed_text = "".join(reversed(text))
print(reversed_text)
```

```python
# using loop
# medium

text = "hello"
rev = ""
for char in text:
    rev = char + rev
print(rev)  # Explanation: each character is added to the front.
```

```python
# using while loop
text = "hello"
rev = ""
i = len(text) - 1

while i >= 0:
    rev += text[i] 
    i -= 1
print(rev)
```

```python
# Using list comprehension

text = "hello"
rev = "".join([text[i] for i in range(len(text) -1, -1, -1)])
print(rev)
```

```python
# Using recursion = when a function call itself to solve a small part of the same problem
# structure = 1.base case (when to stop recursion); 2.recursive case (function calls itself)
# ❌ Without base case → infinite loop → error

# slow

def reverse_string(s):
    # base case
    if len(s) == 0:
        return s
    return reverse_string(s[1:]) + s[0]

reverse_string("hello")
```

```python
# stack = LIFO approach
# slow

text = "hello"
stack = list(text)
rev = ""

while stack:
    rev += stack.pop()

print(rev)
    
```
