```python
#Write a program to check if a number is even or odd.
# num = int(input("enter number: "))
num = 10**18

if num % 2 == 0:
    print(num,"Even Number")
else:
    print(num,"Odd Number")
```

```python
# Ternary operators
num = int(input("enter number: "))
print("even number" if num % 2 == 0 else "odd number")
```

```python
# using dictionary mapping
num = int(input("enter number: "))
result = {0:"Even", 1:"Odd"}[num % 2]
print(result)
```

```python
# using while loop
num = int(input("enter number: "))
temp = num

while temp >= 2:
    temp -= 2
    
if temp == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
# without using modulo ( % )
num = int(input("enter number: "))

if (num // 2) * 2 == num:
    print("Even Number")
    
else:
    print("Odd Number")
```

```python
# using bitwise operator
# Checks last bit (1 = odd, 0 = even)
num = int(input("enter number: "))

if num & 1 == 0:
    print("Even")
else:
    print("Odd")
```

```python
num = float(input("enter number: "))

if num.is_integer():
    num = int(num)
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
else:
    print(num, "is not an integer, so it cannot be odd or even" )
```
