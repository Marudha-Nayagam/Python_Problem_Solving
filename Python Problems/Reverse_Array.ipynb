```python
# reverse() - using buit-in method - (This modifies the original list)
my_list = [1,2,3,4]
my_list.reverse()
print(my_list)
```

```python
# Using slicing
my_list = [1,2,3,4]
reversed_list = my_list[::-1]
print(reversed_list)
```

```python
# Using reversed() function -  (Returns an iterator â€” need list() to convert)
my_list = [1,2,3,4]
reversed_list = list(reversed(my_list))
print(reversed_list)
```

```python
# Reverse using a for loop (prepend method)
my_list = [1,2,3,4]
rev = []
for num in my_list:
    rev = [num] + rev
print(rev)
```

```python
# Reverse using a for loop with indexing
my_list = [1,2,3,4]
rev = []
for num in range(len(my_list)-1,-1,-1):
    rev.append(my_list[num])

print(rev)
```

```python
# Reverse array in-place (swap method)
arr = [1,2,3,4,5,6]
start = 0
end = len(arr) - 1

while start < end:
    arr[start], arr[end] = arr[end], arr[start]
    start += 1
    end -= 1

print(arr)
```
