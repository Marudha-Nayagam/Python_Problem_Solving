```python
# A program processes a list of numbers and needs to find the largest value.

# using built in max()
numbers = [10, 25, 3, 78, 56]
largest = max(numbers)
print(numbers)
print("Largest Number:",largest)
```

```python
# manual method
numbers = [10, 25, 3, 78, 56]
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest Number:",largest)
```
