```python
# Returns a dict_keys view object.

#1.They update automatically if the dictionary changes.

#2. They use less memory.

#3.They allow efficient looping.

    
dictionary = {"a": 1, "b": 2, "c": 3}
keys = dictionary.keys()
print(keys)

dictionary["d"] = 4 # Output updates automatically ------- A normal list would not update like this.
print(keys)
```

```python
# get a list of all the keys in a dictionary
dictionary = {"a": 1, "b": 2, "c": 3}
keys_list = list(dictionary)
print(keys_list)
```

```python
my_dict = {"name": "John", "age": 25, "city": "Chennai"}
keys = list(my_dict.keys())
print("Keys of Dictionary",keys)

# get values
values = list(my_dict.values())
print("Values of Dictionary:",values)

# object that contains (key, value) pairs as tuples
items = list(my_dict.items())
print("Keys and Values in tuple:", items)
```

```python
# using loop
my_dict = {"name": "John", "age": 25, "city": "Chennai"}
keys = []
for key in my_dict:
    keys.append(key)
print(keys)
```

```python
# List comprehension
my_dict = {"name": "John", "age": 25, "city": "Chennai"}
keys = [key for key in my_dict]
print(keys)
```
