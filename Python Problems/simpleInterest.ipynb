```python
# Simple interest
# formula => SI = (P × T × R) / 100

P = float(input("enter principal amount: " ))
T = float(input("enter tennure (in years): "))
R = float(input("enter rate of interest: "))

SI = (P * T * R) / 100
Total = P + SI

print("Simple Interest:", SI)
print("Total amount to be paid:", Total)
```
