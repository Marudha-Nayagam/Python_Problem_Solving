```python
#The Two Sum problem asks us to find two numbers in an array that sum up to a given target value. 
#We need to return the indices of these two numbers.
```

```python
# Approach: Brute force
class Solution:
    def twoSum(nums: list[int], target:int) -> list[int]:
        n = len(nums)
        for i in range(n-1):
            for j in range(i+1,n):
                if nums[i] + nums[j] == target:
                    return [i,j]
        return []
        
```

```python
result = Solution.twoSum([3,2,4],6)
print(result)
```

```python
# Approach: two-pass hash table
# def twoSum(nums:list[int], target:int):
nums = [3,2,4]
n = len(nums)
numMap = {}
target = 6

for i in range(n):
    numMap[nums[i]] = i

for i in range(n):
    complement = target - nums[i]
    if complement in numMap and numMap[complement] != i:
        print([i,numMap[complement]])

# twoSum([3,2,4],6)
```

```python
# Approach: One-pass hash table
nums = [3,2,4]
n = len(nums)
numMap = {}
target = 6

for i in range(n):
    complement = target - nums[i]
    if complement in numMap:
        print([numMap[complement],i])
    else:
        numMap[nums[i]] = i
        
```

```python
# Approach: One-pass hash table - using enumerate
nums = [3,2,4]
n = len(nums)
numMap = {}
target = 6

for i,value in enumerate(nums):
    complement = target - value
    if complement in numMap:
        print([numMap[complement],i])
        break
    else:
        numMap[value] = i
```
