```python
# Body Mass Index
bmi = float(input("Enter the BMI Index:"))
if bmi< 18.5:
    print("Under Weight")
elif bmi < 24.9:
    print("Normal Weight")
elif bmi < 29.9:
    print("Over Weight")
elif bmi < 39.9:
    print("Obese")
else:
    print("Morbidly Obese")
```
